```python
#활용할 변수만 뽑아서 쓰기
column_list = ['matchId','맵이름','리타이어율','accountNo','Speed_Item','Team_Indi','matchRetired']
```

```python
# 유저별로 matchID개수 (각 유저별 게임 횟수) 및 유저의 맵별 플레이 횟수세기
data['play_count_col'] = data.groupby(['accountNo']).matchId.transform('count')
data['playPermap_count_col'] = data.groupby(['accountNo','맵이름']).matchId.transform('count')
```

### 유저별 스개/스팀/아개/아팀 플레이 비율

```python
# data['total_type'] = data['Speed_Item'] + data['Team_Indi']
# #아개(item_solo)
# item_solo = data[data['total_type'] == 0]
# item_solo['total_type_count'] = item_solo.groupby(['accountNo']).matchId.transform('count')
# item_solo_play_col_num = pd.merge(item_solo, data, on = 'accountNo')
```

```python
# 유저별로 스피드전/팀전 플레이 /리타이어 횟수 및 맵별 리타이어 횟수 세기
data['speed_count_col'] = data.groupby(['accountNo']).Speed_Item.transform('sum')
data['team_count_col'] = data.groupby(['accountNo']).Team_Indi.transform('sum')
data['retired_count_col'] = data.groupby(['accountNo']).matchRetired.transform('sum')
data['retiredPermap_count_col'] = data.groupby(['accountNo','맵이름']).matchRetired.transform('sum')
```

```python
#유저별 스피드전/팀전 플레이 개인의 리타이어 비율(생각해보니까 이거는 이미 있네)
data['speed_ratio'] = data['speed_count_col'] / data['play_count_col']
data['item_ratio'] = 1-(data['speed_count_col'] / data['play_count_col'])
data['team_ratio'] = data['team_count_col'] / data['play_count_col']
data['solo_ratio'] = 1-(data['team_count_col'] / data['play_count_col'])
data['retired_ratio'] = data['retired_count_col'] / data['play_count_col']

## 유저의 맵별 리타이어
data['retiredPermap_ratio'] = data['retiredPermap_count_col'] / data['playPermap_count_col']
```

```python
#도전적인 사람의 기준: 개인의 해당 맵에서 유저의 리타이어율 > 맵 리타이어율
final_data = data[['맵이름','play_count_col','accountNo','speed_ratio','item_ratio','team_ratio','solo_ratio','retiredPermap_ratio','리타이어율']]
final_data['challenging'] = final_data['retiredPermap_ratio'] > final_data['리타이어율']
final_data['challenging'] = final_data['challenging'].astype(int)
```

```python
## 164544명의 유저들!! vs 유저별로 플레이한 맵의 고유 행 : 1797776
challenging_ppl = final_data.drop_duplicates(subset=['accountNo','맵이름'])
challenging_ppl
```

```python
#유저의 각 맵별 도전 지수를 합한것(즉, 도전한 맵의 개수)을 도전 점수로 평가
challenging_ppl['challenge_score'] = challenging_ppl.groupby(['accountNo']).challenging.transform('sum')

#각 유저가 사용한 고유 맵의 개수가 많을 수록 위의 도전 지수가 높아질 것을 고려하여 갯수를 고려한 평균값(비율)
#로 계산함. -> 도전지수가 1이므로 평균으로 비율을 계산 가능.
challenging_ppl['challenge_Ratio_score'] = challenging_ppl.groupby(['accountNo']).challenging.transform('mean')
```

```python
A = challenging_ppl['challenge_Ratio_score'].quantile(q=0.25, interpolation='nearest')
B = challenging_ppl['challenge_Ratio_score'].quantile(q=0.5, interpolation='nearest')
C = challenging_ppl['challenge_Ratio_score'].quantile(q=0.75, interpolation='nearest')
C
```

```python
import numpy as np
challenging_ppl['challenge_Ratio_score_categ']  = np.where(challenging_ppl['challenge_Ratio_score'] < A , 0 , 
         np.where(challenging_ppl['challenge_Ratio_score'] < B,1,
                 np.where(challenging_ppl['challenge_Ratio_score'] < C,2,3)))
```
